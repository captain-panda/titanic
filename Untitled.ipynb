```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
from IPython.display import display
pd.options.display.max_columns = None
pd.options.display.max_rows = None
%matplotlib inline

from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.linear_model import SGDClassifier, LogisticRegressionCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RandomizedSearchCV, train_test_split
```

```python
class CateValSeparator(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.df = df
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        d = X.select_dtypes(exclude = np.number)
        return d
```

```python
class UnwantedVal(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.df = df
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        d = X.drop(columns = ['Name', 'Ticket', 'Cabin'])
        return d
```

```python
class Encoder(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.df = df
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        enc1 = LabelEncoder().fit(X['Sex'])
        X['Sex'] = enc1.transform(X['Sex'])
        enc2 = LabelEncoder().fit(X['Embarked'])
        X['Embarked'] = enc2.transform(X['Embarked'])
        return X
```

```python
class EmbarkedFill(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.df = df
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        emb = X['Embarked']
        emb = emb.fillna(emb.value_counts().index[0])
        X['Embarked'] = emb
        return X
```

```python
class NewVal(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.df = df
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        #age
        size = X['Age'].isnull().sum()
        mean = np.around(X['Age'].mean(), decimals = 0)
        std = np.around(X['Age'].std(), decimals = 0)
        random_age = np.random.randint(mean - std, mean + std, size)
        age = X['Age']
        age[np.isnan(age)] = random_age
        X = X.drop(columns = 'Age')
        X['Age'] = age
        
        size = X['Fare'].isnull().sum()
        if(size > 0):
            mode = X['Fare'].mode
            fare = X['Fare']
            fare[np.isnan(fare)] = mode
            X = X.drop(columns = 'Fare')
            X['Fare'] = fare
            
        #new features
        X['Total'] = 1 + X['Parch'] + X['SibSp']
        return X
```

```python
class NumValSeparator(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.df = df
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        d = X.select_dtypes(include = np.number)
        return d
```

```python
df = pd.read_csv('./train.csv', index_col = 'PassengerId')

label = df['Survived']
df = df.drop(columns = 'Survived')

num_pipe = Pipeline([
    ('numval', NumValSeparator()),
    ('newval', NewVal())
])

cat_pipe = Pipeline([
    ('catval', CateValSeparator()),
    ('unwanted', UnwantedVal()),
    ('embarkedfill', EmbarkedFill()),
    ('encoder', Encoder())
])


l = num_pipe.fit_transform(df)
k = cat_pipe.fit_transform(df)
new = pd.concat([l, k], axis = 1)

rfc = RandomForestClassifier(oob_score = True, n_jobs = 1, max_samples = 0.8)

#randomized search cv
random_grid = {
    'n_estimators': np.arange(200, 2000, 10),
    'max_depth': np.arange(5,10,5)
}
randomized = RandomizedSearchCV(rfc, random_grid, random_state = 42, n_iter = 10)
randomized.fit(new, label)

best = randomized.best_estimator_
'''
best.fit(new, label)
print(best.oob_score_)
'''
```
